p = $\infty$:

For en diagonalmatrise er altså normene p = 1,2 og $\infty$ den samme!

```python
import numpy as np
import numpy.linalg as la
from scipy.linalg import hilbert
from scipy.linalg import invhilbert #denne lager eksplisitt den inverse av hilbertmatrisen
# bvec er en liten funksjon som gir oss vector med n 1'ere.
bvec= lambda n: np.ones((n,))
# print(bvec(3))
# Kondisjonstallet til 4x4 Hilbertmatrisen
# print(la.cond(hilbert(4)))

# Lag en perturbasjon av norm epsilon
def b_perturb(n,epsilon):
    d = 2*np.random.rand(n,1)-1 #genererer random tall fra -1 til 1
    d = d/la.norm(d,2)*epsilon #første del normaliserer delta. Epsilon må da skalere den til ||d|| = 0.01 ||b||
    return d

"""Oppgave a"""

def solver(n,rand = True,better_b = False):
    H = hilbert(n)
    b = bvec(n)
    if better_b:
        for j in range(20):
            b = H@b
            scale = la.norm(b,2)
            b = b/scale
    x = la.solve(H,b)
    maxerror = 0
    epsilon = 0.01*la.norm(b,2)
    for i in range(1000):
        if rand:
            delta = b_perturb(n,epsilon)
        else:
            delta = np.array([0,0.012,-0.016,0])
        y = la.solve(H,b+delta)
        error = la.norm(y-x,2)
        if error > maxerror:
            maxerror = error
    K_estimate = maxerror / la.norm(x,2) * la.norm(b,2)/la.norm(delta,2)
    return K_estimate
Estimates =[]
K_conds =[]
for n in (5,10,15,20):
    Estimates.append(solver(n))
    K_conds.append(la.cond(hilbert(n)))
K_estimates = np.array(Estimates)
K_reals = np.array(K_conds)
print("K estimates \t K reals")
for i in range(4):
    print(f"{K_estimates[i]:.2e}\t{K_reals[i]:.2e}")
    
Kontroll = solver(4,rand = False)
print(f'Kontrollspørsmål har svaret: {Kontroll}')


""""Oppgave b""" 

b_Estimates =[]
for n in (5,10,15,20):
    b_Estimates.append(solver(n,better_b = True))
b_K_estimates = np.array(b_Estimates)
print("K estimates \t K reals")
for i in range(4):
    print(f"{b_K_estimates[i]:.2e}\t{K_reals[i]:.2e}")
```
