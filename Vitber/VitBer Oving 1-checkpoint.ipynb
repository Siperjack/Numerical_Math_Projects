**Oppgave 2**

La oss forsøke å gjøre noe "fornuftig". Så langt har vi i timene lært om intervallhalveringsmetoden, så la oss lage en enkel implementasjon av denne. Vi søker nullpunkter i funksjonen
$$
f(x) = \cos (x^2) - \sin x
$$

**(a)** Start med å lage en Python-funksjon som definerer $f(x)$ og plott den i intervallet $[-\pi,\pi]$.

**Kontrollspørsmål 1** Hvor mange nullpunkter har $f(x)$ i dette intervallet sett ut fra grafen?


```python
def f(x):
    # Implementer f som definert over
    return np.cos(x**2)-np.sin(x)
    
# Skriv kode som plotter funksjonen f(x) mellom -pi og pi. Du får bruk for np.linspace.
x_values = np.linspace(-np.pi,np.pi,100)
y_values = [f(x) for x in x_values]
plt.title("f(X) = cos(x^2)-sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x_values,y_values)
plt.show()
```

**(b)** Skriv en funksjon med navn *inthalv* som tar 4 argumenter:
* $f$: Funksjonen man skal finne nullpunkter av.
* $a$: Venstre endepunkt i startintervallet.
* $b$: Høyre endepunkt i startintervallet (må ha $f(a)\cdot f(b) < 0$).
* tol: En toleranse slik at $|r-m|<\mathrm{tol}$ der $r$ er roten og $m$ er midtpunktet av det siste beregnede intervallet.

Funksjonen skal returnere midtpunktet $m$ i det sist beregnede intervallet, samt antall halveringer som er gjort.
Kjør så funksjonen med $f$ som ovenfor, $a=0$, $b=\frac{\pi}{2}$, $\mathrm{tol}=10^{-7}$.

**Kontrollspørsmål 2** Er roten du får ut med disse inngangsverdiene større eller mindre enn $0.84937$ ?

**Kontrollspørsmål 3** Hvor mange intervallhalveringer bruker funksjonen for å finne svaret?

```python
def inthalv(f,a,b,tol):
    # Fjern pass og fullfør funksjonen
    count = 0
    while abs(b-a)/2 > tol:
        count+=1
        c = (a + b)/2
        if f(c) == 0:
            return c
        elif f(a)*f(c)<0:
            b = c
        else:
            a = c
    print("Antall halveringer er: ", count)
    return (b + a)/2
        


# Skriv kode som definerer input og kaller inthalv her
my_tol = 0.0000001
root_f = inthalv(f,0,np.pi/2, my_tol)
print(f"The root found is between -3 and -2 with tolerance {my_tol} is: {root_f:.6f}, and put into f this equals to: {f(root_f):.5f}")
```
