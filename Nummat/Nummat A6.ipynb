## Problem 1

Determine a quadrature formula
\begin{equation*}
  I_2(f) = w_1f(0)+w_2f(\pi),
\end{equation*}
which integrates functions of the form
\begin{equation*}
  f(x) = a+b\cos(x),\quad a,b\in \mathbb R, 
\end{equation*}
__exactly__ when using the interval $[0,2\pi]$.
Show that such a quadrature formula is also exact for all functions of the form
\begin{equation*}
  f(x) = \sum_{k=0}^n a_k\cos((2k+1)x)+b_k\sin(kx), \quad a_k,b_k\in \mathbb R.
\end{equation*}

**Solution 1**

We create a system of equations for the weight, where the first row is the equation to satisfy solving the integral of the constant function exactly with the quadrature, the second for $\cos(x)$. Here
$$
\left[\begin{matrix} 1& 1\\\cos(0)& \cos(\pi) \\\end{matrix}\right]\left[\begin{matrix} w_1 \\ w_2 \end{matrix}\right]
= \left[\begin{matrix} \int_0^{2\pi}{1}dx \\ \int_0^{2\pi}{\cos(x)}dx\end{matrix}\right]
$$
Evaluating the functions and integrals we get
$$
\left[\begin{matrix} 1& 1 \\1 & -1\end{matrix}\right]\left[\begin{matrix} w_1 \\ w_2\end{matrix}\right]
= \left[\begin{matrix} 2\pi \\ 0\end{matrix}\right]
$$

From this we can easly read of that the weights must be
$$
\left[\begin{matrix} w_1 \\ w_2 \end{matrix}\right] = \left[\begin{matrix} \pi \\ \pi \end{matrix}\right]
$$

Because the quadrature with these weights will solve the integrals of the basisfunctions exactly, all linear combinations will be solved exactly too.

We then look at \begin{equation*}
  f(x) = \sum_{k=0}^n a_k\cos((2k+1)x)+b_k\sin(kx), \quad a_k,b_k\in \mathbb R.
\end{equation*}

or particularly \begin{equation*}
  f_k(x) = a_k\cos((2k+1)x)+b_k\sin(kx), \quad a_k,b_k\in \mathbb R.
\end{equation*}

For this to be evaluated exactly by a quadrature needs to evaluate to zero, as the integral of a trigonometric function over a period must be zero. We then get the equation $$
\left[\begin{matrix} f_k(0) & f_k(\pi) \end{matrix}\right]\left[\begin{matrix} w_1 \\ w_2\end{matrix}\right] = 0
$$
which evaluates to $$
\left[\begin{matrix} 1 & -1 \end{matrix}\right]\left[\begin{matrix} w_1 \\ w_2\end{matrix}\right] = 0
$$
As this is the only criteran for evaluating the function $f_k(x)$ exactly, and this is fulfilled with the weights already computed in the first part, so $f(k)$ as will therefore be solved exactly as all its terms are solved exactly.

## Problem 2

Compute the Fourier coefficients  $c_k = (f, \mathrm{e}^{2\pi i k \cdot}) = \displaystyle\int_{-\frac{1}{2}}^{\frac{1}{2}} f(x) \mathrm{e}^{-2\pi i k x} \mathrm{d} x$, $k\in\mathbb Z$ of the following functions:

1. $f_1(x) = 4\cos(2\pi\ 9x)$
2. $f_2(x) = 6\sin(2\pi\ 9x)$
3. $f_3(x) = \bigl(\cos(2\pi x) - \sin(2\pi x)\bigr)^3$
4. $f_4(x) = \begin{cases} 1 &\text{ if } x \in [-a,a), \\ 0 & \text{ otherwise.} \end{cases}$

__Hint__: The coefficients of the first three functions can be computed using Euler's identity instead of computing an integral.

Compute now also the Fourier coefficients $a_0, a_k, b_k$, $k=1,\ldots,$.

**solution**
1. $f_1(x) = 4\cos(2\pi\ 9x)$
$$c_k = \begin{cases} 2 & \text{if} & k = \pm 9 \\ 0 & \text{else} \end{cases}$$
2. $f_2(x) = 6\sin(2\pi\ 9x)$
$$c_k = \begin{cases} \pm 3i & \text{if} & k = \pm 9 \\ 0 & \text{else} \end{cases}$$
3. $f_3(x) = \bigl(\cos(2\pi x) - \sin(2\pi x)\bigr)^3$
$$c_k = \begin{cases} \sqrt{2}\frac{-1 \pm i}{2}& \text{if} & k = \pm 3 \\  \sqrt{2}\frac{3 \mp 3i}{2} & \text{if} & k = \pm 1\\ 0 & \text{else}\end{cases}$$
4. $f_4(x) = \begin{cases} 1 &\text{ if } x \in [-a,a), \\ 0 & \text{ otherwise.} \end{cases}$

Here $c_0$ =$2a$ as this is trivially the avg value of the funciton over the entire period. Evaluating the integral we get 
$$c_k = \begin{cases} \frac{\sin{2\pi k a}}{\pi k}\text{if} & k \neq 0 \\ 2a & \text{if} & k = 0\end{cases}$$
We here assume $\lvert a \rvert \leq \frac{1}{2}$

## Problem 3

Implement the explicit Euler method. Your function should look like

```
 y = Euler(f, t0, T, h, y0)
```

where $f$ is a function handle for the right hand $f(t,y(t))$ of the corresponding initial value problem.

Test your function with the predator-prey model (Vottka-Volterra)
\begin{align}
  x'(t) &= \alpha x(t) - \beta x(t)y(t)\\
  y'(t) &= \delta x(t)y(t) - \gamma y(t)
\end{align}
with the following initial values and starting parameters:
\begin{align*}
  y_0 = (100,10), \qquad T=20, \qquad \alpha = 1, \qquad \beta = \delta = 0.05, \qquad \gamma = 2.
\end{align*}
Plot both components into one plot with a common time axis.
Choose a proper $h$.

Compare the result to an own implementation of Heun's method. _Alternatively_: Find an existing implementation of Heun's Method in the defaults _or_ look for a suitable implementation.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def Vottka_Volterra(t, x, alpha = 1,beta = 0.05, delta = 0.05, gamma = 2):
    return np.array([alpha*x[0] - beta*x[0]*x[1], delta*x[0]*x[1] - gamma*x[1]])
````

````python
def euler(f, t0, T, h, y0):
    N = int(np.ceil((T - t0)/h))
    t = np.linspace(t0, T, N)
    y = np.zeros((N, len(y0)))
    y[0] = y0
    for i in range(N - 1):
        y[i + 1] = y[i] + h * f(t[i], y[i])
    return y, t  
````

````python
y0 = np.array([1000, 1])
t0 = 0
T = 100
h = 0.01
````

````python
y, t = euler(Vottka_Volterra, t0, T, h, y0)
````

````python
plt.title("x(t)")
plt.plot(t, y[:, 0])
````

````python
plt.title("y(t)")
plt.plot(t, y[:, 1])
````

````python
plt.title("Phasespace)")
plt.plot(y[:, 0], y[:, 1])
````
