1. Compute the gradient $\nabla F$ and the Hessian $H_F$ of the Rosenbrock function $F\colon \mathbb R^2 \to \mathbb R$ defined by $F(\mathbf x) = 100(x_2-x_1^2)^2 + (1-x_1)^2$.
   Show that $\mathbf x^* = (1,1)^\mathrm{T}$ is the only local minimizer of this function and that the Hessian matrix at this point is positive definite.

##### Function description: computeGradAndHess
computeGradAndHess takes in an array of arguments of F, together with and the scalarfunction F, and then for each component of x computes the gradient and hessian. This function assumes that F i twice differentiable in respect to all arguments. 

```python
def computeGradAndHess(x,F):
    N = len(x)
    grad_F = smp.zeros(N,1)
    hess_F = smp.zeros(N,N)
    for i in range(N):
        grad_F[i] = smp.diff(F,x[i]) #array of partial derivatives for given function F
    for i in range(N):
        for j in range(N):
            hess_F[i,j] = smp.diff(grad_F[i],x[j]) #2Darray of 2nd order partial derivatives for given function F
    return grad_F, hess_F
```

Defining F and G as python functions

```python
def F(x):
    return 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2

def G(x):
    return 8*x[0] + 12* x[1] + x[0]**2 - 2*x[1]**2
```

Statement: The point $x* = [1,1]$ is a local stabilizer of F, and it is also the only one. 

Proof: 

Taking the trace and determinant equaling 1000 and 400, the characteristic polyomial $$ p(\lambda) = \lambda^2 -trace \lambda + det $$ gives $$ \approx \lambda = {999.6,0.4} $$ which are both positive, which is sufficient to say the hessian matrix is positive definite given that it is symmetric. We also see that only $x1 = \pm 1$ and $x2 =1$ allows the second row of the gradient to be evaluated at zero. However only $x1 = x2 = 1$ evaluate to zero in the first row. This means $x* = [1,1]$ is the only stationary point in F, and the only candidate to be a local minimizer. The positiv definite evaluated hessian matrix in this point then guaranties that this point is no saddlepoint, and this is sufficient to conclude that $x*$ is a local minimizer of F, and as it is the only point evaluating the gradient to zero, it is unique.

Q.E.D.

## Problem 3

1. Implement a generic (simple) Newton iteration function of the form

       def my_newton(f, Df, x0, **kwargs)

   To look for a zero of the univariate function $f\colon \mathbb R \to \mathbb R$.
   Hence the function performs the newton iteration for `f` and its derivative `Df` starting from `x0` performing `maxiter` steps, where `maxiter` is a keyword argument from `**kwargs`.
   Provide a suitable default if no `maxiter` keyword is given.

2. Extend the function from the first part of this problem with an additional keyword `tol` to stop, if $\lvert f(x_k)\rvert$ is smaller in absolute value than the tolerance `tol`.
   What is a good default here to make the function work the same as in 1, i.e. that the default behavior (from 1) is not changed by this additional keyword?

3. Document the function properly.

4. Write at least two tests with simple functions. One should end with `maxiter` iterations, the other with a tolerance stop.


```python
def my_newton(f,Df,x0,maxiter = 1000, tol = 1E-7, printing = True): #maxiter = 1000, tol = 1E-7 seems resonable as default stopping critera
    ###Initialising some variables for first cycle
    error = 10 #Just so the first cycle will run. Choosing tol>10 will just give tol = 10 and not run any iterations.
    count = 0
    x = x0
    
    while count < maxiter and error > tol:
        count += 1
        ###Uppdating x(k) with its next iteration x(k+1)
        
        x -= f(x)/Df(x) #The uppdating rule in newton iteration

        error = np.linalg.norm(f(x)) #The choice of errorfunction of the task
                             
    if count == maxiter:
        if printing:
            print(f"maxiter reached at {maxiter} reached") #Prints to screen if the stoppingcritera reached is maxiter instead
        # of a solution inside our allowed toleranse bounds.
    return x,count,error
```

An example function and its derivative used as testcases:

```python
def fex(x):
    return x**4 - 2*x**2 + 2*x - 2
def fex_d(x):
    return 4*x**3 - 4*x + 2
```

In the codeblock below we both times choose $x_0 = 1$, but the secound call we only allow 1 iteration, which should print "maxiter reached" to screen, and the a solution outside of the tolerance.

```python
print("Case 1(low maxiter): ")
sol, count, error = my_newton(fex,fex_d,1)
print(sol)
print("count: ",count)
print(f"error is {error:.2e}")# \nsol is {x:.2e}")
print("\nCase 1(high maxiter): ")
sol, count, error = my_newton(fex,fex_d,1,maxiter = 1)
print(sol)
print("count: ",count)
print(f"error is {error:.2e}")# \nsol is {x:.2e}")
```

##### 4.1)
To find the solution to an equality with newtons method, we write it in normal form with an expression at one side and 0 at the other. We then call this expression f(x). In this case we get: $$ z^3 = 1 <=> z^3 - 1 = f(z) = 0$$
The fundamental theorem of algebra then tells us this polynomial of degree 3 got 3 roots $r_i \in \mathbb{C}$

```python
def f_z_1(z):
    return z**3 - 1
def Df_z_1(z):
    return 3*z**2
```

```python
def eval_newton(f,Df,n,maxiter = 100, tol = 1E-7, printing = True):
    A = np.zeros((n,n),dtype = complex)
    x = np.linspace(-1,1,n)
    y = np.linspace(-1j,1j,n)
    for i in range(n):
        for k in range(n):
            aikz,count,error  = my_newton(f,Df,x[i]+y[k], maxiter = maxiter, tol = tol,  printing =printing)
            A[i,k] = aikz
    return A
```

```python
sols1 = eval_newton(f_z_1,Df_z_1,n=256)
```
